EXERCICE : importer et charger une image puis zoomer et contraster

```python
from scipy import misc
```

```python
import matplotlib.pyplot as plt
```

```python
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
type(face)
```

```python
face.shape
```

```python
# recadrer et zoomer sur la photo
h = face.shape[0]
l = face.shape[1]
hzmin = round(h*0.25)
hzmax = round(h*0.75)
lzmin = round(l*0.25)
lzmax = round(l*0.75)
print(hzmin, hzmax, lzmin, lzmax)
```

```python
face = face[hzmin:hzmax, lzmin:lzmax]
face.shape
```

```python
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
# luminosité max (255 = blanc) pour tous les pixels > 200
face[face>200] = 255
# luminosité min (0 = noir) pour tous les pixels < 50
face[face<50] = 0
```

```python
# résultat : une image plus contrastée
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```
